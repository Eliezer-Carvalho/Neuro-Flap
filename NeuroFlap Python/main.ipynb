```python
import pandas as pd
import matplotlib.pyplot as plt
```

<h3> <b> ReLU -> Sigmoid </b> </h3>

```python
df = pd.read_csv(r"C:\Users\User\Desktop\NeuroFlap Python\ReLU_Sigmoid\ReLU_Sigmoid.csv")

df.value_counts()

PRIMEIRA_GERAÇÃO_VENCEDOR = 16
NÚMERO_VENCEDORES = df[df == 6780].count().values.sum()

MÉDIA = df.mean(axis = 0).sum().round()

TEMPO = 50 #Minutos

display (f"Geração que atingiu o objetivo a primeira vez -> {PRIMEIRA_GERAÇÃO_VENCEDOR}")
display (f"Média Total de Fitness -> {MÉDIA}")
display (f"Tempo de Execução -> {TEMPO} minutos")
display (f"Quantidade de Indivíduos Vencedores -> {NÚMERO_VENCEDORES}")
```

```python
MÉDIA_COLUNA = df.mean(axis = 0)

x = list(range(1, 51, 1))
y = MÉDIA_COLUNA.values

plt.style.use('bmh')
plt.figure(figsize = (34, 10)) #Largura x Altura
plt.plot(x, y, label = "teste", color = "navy", marker = ".")


plt.xlabel ("Gerações", size = 25, labelpad = 12)
plt.xticks(list(range(1, 52, 1)))
plt.ylabel ("Média do Fitness", size = 25, labelpad = 12)
plt.title ("ReLU -> Sigmoid", size = 25, pad = 20)

plt.show()
```
